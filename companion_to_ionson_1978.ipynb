Equation 5 - Conservation of mass

$\frac{\partial \rho}{\partial t} + \nabla \left(\rho v \right) = 0$

Equation 6 - Momentum conservation

$\frac{d v}{d t} = \frac{J \times B}{4 \pi \rho} - \frac{\nabla P}{\rho}$

Equation 7 - Ideal Ohm's law

$E + \frac{v}{c} \times B=0$

Equation 8 - Ampere's law

$\nabla \times B = \frac{4 \pi}{c} J$

Equation 9 - Faradays's law

$\frac{1}{c}\frac{\partial B}{\partial t} = -\nabla \times E$

Equation 10 - Ideal gas law

$P = n_e k T_e$

Equation 11 - Eliminating the current density from the momentum equation using Ampere's law

$4 \pi \rho \frac{d v}{d t} = \left(\nabla\times B\right) \times B - 4 \pi \nabla P$

Equation 12 - Eliminiating the electric field from Faraday's equation using the Ideal Ohm's law

$\frac{\partial B}{\partial t} = \nabla \times \left(v \times B\right) = \left(B \cdot \nabla \right) v - \left(v\cdot \nabla \right) B - B \left(\nabla \cdot v \right)$

In steady state $\partial v / \partial t = \partial B / \partial t = 0$. The author introduces a new variable: the displacement, which relates to the velocity through $v = d \xi / d t$. Linearising the momentum equation yields equation 13

$4 \pi \rho \frac{\partial^2 \xi}{\partial t^2} = \left(\nabla \times B'\right) \times B + \left(\nabla \times B\right) \times B' - 4 \pi \nabla P'$

Equation 14 can be obtained from equation 12 by replacing the velocity with the displacement

$B' = \left(B \cdot \nabla \right) \xi - \left(\xi \cdot \nabla \right) B - B \left(\nabla \cdot \xi\right)$

To derive equation 15, we start from the perturbed conservation of mass

$\rho' = -\xi \nabla \rho - \rho \nabla \xi$

Substituting in the ideal gas equation yields

$\frac{P'}{P} = \frac{\rho'}{\rho} = -\xi \frac{\nabla \rho}{\rho}- \nabla \xi = - \chi_p \xi - \nabla \xi$

Equations 16 - 18

$\left(\nabla \times B'\right) \times B + \left(\nabla \times B\right) \times B' = \left( B \cdot \nabla \right) B' + \left(B' \cdot \nabla \right) B - \nabla \left(B \cdot B' \right)$

We only need to expand the first term

$ \left(B \cdot \nabla \right) B'= \left(B \cdot \nabla \right) \left[ \left(B \cdot \nabla \right) \xi - \left( \xi \cdot \nabla \right) B - B \left( \nabla \cdot \xi\right) \right]$

The remainder of the derivation is straightforward

Derivation of equation 25

```python
epsilon = sympy.Symbol('epsilon')
xi_parallel = sympy.Symbol(r'\xi_{\parallel}')
xi_perp = sympy.Symbol(r'\xi_{\perp}')
k_perp = sympy.Symbol(r'k_{\perp}')
k_parallel = sympy.Symbol(r'k_{\parallel}')
tilde_P_prime = sympy.Symbol(r"\tilde{P}'")
dxirdxr = sympy.Symbol(r'd \xi_r/d x_r')
B = sympy.Symbol('B')
eqn_22 = sympy.Eq(epsilon*xi_parallel, 4*sympy.pi*sympy.I*k_parallel*tilde_P_prime+
                 sympy.I*B**2*k_parallel*(sympy.I*k_parallel*xi_parallel+sympy.I*k_perp*xi_perp+dxirdxr))
eqn_22
```

```python
P = sympy.Symbol('P')
eqn_24 = sympy.Eq(tilde_P_prime/(P+B**2/4/sympy.pi),
                 - dxirdxr-sympy.I*k_perp*xi_perp-sympy.I*k_parallel*xi_parallel*(P/(P+B**2/4/sympy.pi)))
eqn_24
```

```python
eqn_21 = sympy.Eq(epsilon*xi_perp,
                 4*sympy.pi*sympy.I*k_perp*tilde_P_prime)
eqn_21
```

```python
alpha = sympy.Symbol('alpha')
rho = sympy.Symbol('rho')
omega = sympy.Symbol('omega')
eqn_26 = sympy.Eq(alpha, 1+2*sympy.pi*rho*beta*omega**2/(4*sympy.pi*rho*omega**2-beta*k_parallel**2*B**2/2))
eqn_26
```

```python
eqn_23 = sympy.Eq(epsilon, 4*sympy.pi*rho*omega**2-k_parallel**2*B**2)
eqn_23
```

```python
beta = sympy.Symbol('beta')
temp = eqn_24
temp = temp.subs(sympy.solve(eqn_22,xi_parallel,dict=True)[0])
temp = temp.subs(sympy.solve(eqn_21,tilde_P_prime,dict=True)[0])
temp = temp.subs(P, beta*B**2/8/sympy.pi)
temp = temp.subs(sympy.solve([eqn_23, eqn_26], [rho, k_parallel], dict=True)[1])
temp = sympy.solve(temp, xi_perp)[0].simplify()
eqn_25 = sympy.Eq(xi_perp, temp)
eqn_25
```
